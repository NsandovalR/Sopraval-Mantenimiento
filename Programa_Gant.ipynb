```python
from dash import Dash, html, dash_table, dcc, callback, Output, Input
import pandas as pd
import plotly.express as px
from datetime import datetime, timedelta
import tkinter as tk
from tkinter import simpledialog
def strip1(x):
    return(x.strip())
def to_date(x):
    x = datetime.strptime(x ,'%d.%m.%Y')
    return(x)
def to_str(x):
    x = datetime.strftime(x, '%Y-%m-%d')
    return(x)
def df_clipboard(Pasar_a_Fecha = False , Nombre_col_Fecha ='  Fecha ref.',Quitar_cols= False, Nombre_cols_borrar = None  ): #Crea un df, desde una tabla SAP copiada al portapapeles
    df = pd.read_clipboard( engine = 'python' ,sep='|', header = 3 , skiprows = [4], quoting=3,skipfooter=1) 
    df = df.iloc[:, 1:-1]
    if Pasar_a_Fecha == True:
     print("Fecha=True , cambiando formato de fecha SAP a fecha excel (Recuerda adaptarlo en documento)")  
     df[Nombre_col_Fecha] = df[Nombre_col_Fecha].apply(to_date)
    if Quitar_cols == True:
     df = df.drop(columns = Nombre_cols_borrar)
    return(df)
# Función para mostrar el cuadro de diálogo para ingresar credenciales
def ingresar_credenciales():
    username = simpledialog.askstring("Ingresar Credenciales", "Usuario:")
    credenciales.append(username)
    password = simpledialog.askstring("Ingresar Credenciales", "Contraseña:", show="*")
    credenciales.append(password)
    Desde = simpledialog.askstring("Filtro de fechas", "Fecha inicial formato SAP:")
    credenciales.append(Desde)
    Hasta = simpledialog.askstring("Filtro de fechas", "Fecha final formato SAP:")
    credenciales.append(Hasta)
    # Verificar si se ingresaron tanto el usuario como la contraseña
    if username and password:
        print("Credenciales añadidas:")
        root.destroy()
    else:
        print("No se ingresaron credenciales.")
        root.destroy()
credenciales = []
root = tk.Tk()
btn_ingresar = tk.Button(root, text="Ingresar Credenciales", command=ingresar_credenciales)
btn_ingresar.pack()
root.mainloop()
User = credenciales[0]
Password = credenciales[1]
Desde = credenciales[2]
Hasta = credenciales[3]
```
